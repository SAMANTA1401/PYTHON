```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Define the data points
blue_points = np.array([[1, 2], [2, 1], [1, -2], [2, -2]])
red_points = np.array([[4, -1], [4, 1], [5, -1], [6, 1]])

# Create labels for blue (class 0) and red (class 1)
labels = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Train the SVM classifier
clf = svm.SVC(kernel='linear')
clf.fit(np.vstack((blue_points, red_points)), labels)

# Get the support vectors
support_vectors = clf.support_vectors_

# Get the coefficients of the separating hyperplane
w = clf.coef_[0]
b = clf.intercept_[0]

# Plot the data points
plt.scatter(blue_points[:, 0], blue_points[:, 1], color='blue', label='Blue')
plt.scatter(red_points[:, 0], red_points[:, 1], color='red', label='Red')

# Plot the support vectors
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], color='green', marker='x', s=100, label='Support Vectors')

# Plot the optimal separating line
x = np.linspace(0, 7, 10)
y = (-w[0] / w[1]) * x - (b / w[1])
plt.plot(x, y, linestyle='--', color='black', label='Optimal Separating Line')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the data
blue_x = [1, 2, 1, 2]
blue_y = [2, 1, -2, -2]

red_x = [4, 4, 5, 6]
red_y = [-1, 1, -1, 1]

# Plot the data
plt.scatter(blue_x, blue_y, color='blue', label='Blue')
plt.scatter(red_x, red_y, color='red', label='Red')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Red and Blue Categories')
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.svm import SVC

# Create the SVM classifier
clf = SVC()

# Fit the classifier to the data
clf.fit(blue_x + red_x, blue_y + red_y)

# Get the support vectors
support_vectors = clf.support_vectors_

# Get the optimal separating line
w = clf.coef_[0]
b = clf.intercept_[0]

# Calculate the equation of the optimal separating line
def optimal_separating_line(x):
  return (-b - w[0] * x) / w[1]

# Plot the support vectors and optimal separating line
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], color='black', marker='o', label='Support vectors')
plt.plot(blue_x + red_x, optimal_separating_line(blue_x + red_x), color='black', label='Optimal separating line')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Support Vectors and Optimal Separating Line')
plt.legend()

# Show the plot
plt.show()
```
