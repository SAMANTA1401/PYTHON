```python
# : Concatenate two data frames using the following conditions
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 133453,53243251]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995,352352,465346,3423235]}
```

```python
germaancar=pd.DataFrame.from_dict(GermanCars)
```

```python
germaancar
```

```python
japaneseCar=pd.DataFrame.from_dict(japaneseCars)
japaneseCar
```

```python
car = pd.concat([germaancar,japaneseCar],keys=['germany','japan'])
```

```python
car
```

```python
# : Merge two data frames using the following condition\
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 13592, 6099]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80,90, 150] }
```

```python
carprice = pd.DataFrame.from_dict(Car_Price)
carhor = pd.DataFrame.from_dict(car_Horsepower)
```

```python
carsdf = pd.merge(carprice, carhor, on='Company')
carsdf
```
